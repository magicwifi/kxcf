```python
# coding:utf-8
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split


%matplotlib inline
```

```python
train_big_table = pd.read_csv('train_big_table.csv',encoding='utf-8',index_col = False, delimiter='\t')
train_big_table.head()
```

```python
def is_edu_ok(x):
    if x["cur_degree_id"] >= x["min_edu_level"] :
        return True;
    else:
        return False;


train_big_table = train_big_table[train_big_table.apply(is_edu_ok,axis=1)&(train_big_table['work_years_statisfied']==1)]

train_big_table.shape
```

```python
def concat_personal_str(x):
    return str(x['desire_jd_type_id'])+';'+ str(x['cur_jd_type']) +';'+ str(x['experience'])


train_big_table['personal_str'] =  train_big_table.apply(concat_personal_str,axis=1)
```

```python
def concat_jd_str(x):
    return str(x['jd_title']) +';'+ str(x['jd_sub_type'])+ ';' +str(x['job_description'])


train_big_table['jd_str'] =  train_big_table.apply(concat_jd_str,axis=1)
```

```python
train_big_table[['jd_str','personal_str','mix_sat']].to_csv("train_bert.csv",index=None,sep='\t')
```
